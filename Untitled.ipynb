```python
data_small = data_all[data_all['days']<80].reset_index(drop=True)
```

```python
def ols_all(gr):
    gr_small = gr[gr['days'].isin(range(20,80))]
    olsmod = sm.OLS(endog = gr_small['COUNT_norm'], exog = sm.add_constant(gr_small['days']))
    olsres = olsmod.fit()
    return olsres.predict(sm.add_constant(gr['days']))    

data_small['COUNT_int'] = data_small[['CFO', 'days', 'COUNT_norm']].groupby(['CFO'], group_keys = False).apply(
    lambda gr: ols_all(gr))
```

```python
data_small['COUNT_wq'] = data_small['COUNT_norm'] / data_small['COUNT_int']
```

```python
ax = data_small[data_small['days']<45].boxplot(column='COUNT_wq', by='days', figsize = (14, 8))

ax.set_ylim([0,5])
```
